```python
import random

sum = 0
date = 1
flag = True

while flag:
    patientCount = random.randint(50, 100)  # 랜덤 
    sum += patientCount
    date += 1
    print('날짜: {}, \t 오늘 환자 수: {} \t 누적 환자수:{}'.format(date, patientCount, sum))

    if sum >= 10000:                          # while 문 안에서 if 문 넣어준다
        flag = False
```

```python
cnt = 0
for i in range(100): # 
    if i % 7 != 0:
        continue
    
    print('{}는 7의 배수입니다.'.format(i))
    cnt += 1

else:  # 이 엘스는 반복문이 종료된 후에 실행
    print('99까지의 정수 중 7의 배수는 {}개 입니다.'.format(cnt))
```

```python
minNum = 0

for i in range(1, 101):
    if i % 3 != 0 or i % 7 != 0:
        continue
    print('공배수: {}'.format(i))

    if minNum == 0:
        minNum = i

else:
    print('최소 공배수 : {}'.format(minNum))
```
